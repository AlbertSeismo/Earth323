###Direct wave

\begin{equation}
t_1\,=\,\frac{x}{V_1}
\end{equation}

###Refraction

For $x\,<\,x_c$ (critical distance), there is no head wave:
<br>
<br>
\begin{equation}
x_c
\,=\,\frac{2hV_1}{\sqrt{{V_2}^2\,-\,{V_1}^2}}
\end{equation}
<br>
<br>
When $x_c\,\leq\,x\,\leq\,x_c\,+\,x_2$:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{2\sqrt{{x_1}^2\,+\,h^2}}{V_1}
\,+\,\frac{x\,-\,2x_1}{V_2}
\end{equation}
<br>
<br>
Thus:
<br>
<br>
\begin{equation}
\sqrt{{x_1}^2\,+\,h^2}
\,=\,\frac{h}{\cos{i_c}}
\,\implies\,t_2
\,=\,\frac{2h}{V_1\cos{i_c}}
\,+\,\frac{x}{V_2}
\,-\,\frac{2h\sin{i_c}}{V_2\cos{i_c}}
\end{equation}
<br>
<br>
Therefore:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{x}{V_2}
\,+\,\frac{2h}{V_1}\left[\frac{1\,-\,\displaystyle\frac{V_1}{V_2}\sin{i_c}}{\cos{i_c}}\right]
\end{equation}
<br>
\begin{equation}
1\,-\,\frac{V_1}{V_2}\sin{i_c}\,=\,{\cos{i_c}}^2
\,\implies\,t_2
\,=\,\frac{x}{V_2}
\,+\,\frac{2h\cos{i_c}}{V_1}
\end{equation}
<br>
<br>
Hence:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{x}{V_2}
\,+\,\frac{2h\sqrt{{V_2}^2-{V_1}^2}}{V_1V_2}
\end{equation}
<br>
<br>
Being $l\,=\,x_1\,+\,x_2$, for $x\,\geq\,l\,+\,x_4$:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{\sqrt{{x_1}^2\,+\,h^2}}{V_1}
\,+\,\frac{x_2}{V_2}
\,+\,\frac{d}{V_2}
\,+\,\frac{\sqrt{{x_4}^2\,+\,\left(h\,-\,h'\right)^2}}{V_1}
\end{equation}
<br>
<br>
\begin{equation}
x_2
\,=\,l\,-\,\frac{x_c}{2}
\text{,}\qquad
d
\,=\,\frac{x_3}{\cos{\phi}}
\text{,}\qquad
h'
\,=\,x_3\tan{\phi}
\qquad\text{e}\qquad
x_4
\,=\,(h\,-\,h')\tan(i_c\,-\,\phi)
\end{equation}
<br>
<br>
\begin{split}
\sqrt{{x_4}^2\,+\,(h\,-\,h')^2}
&=\,\sqrt{\left[(h\,-\,h')\tan(i_c\,-\,\phi)\right]^2\,+\,(h\,-\,h')^2}\\
&=\,(h\,-\,h')\sqrt{1\,+\,\tan(i_c\,-\,\phi)^2}
\,=\,\frac{h\,-\,x_3\tan{\phi}}{\cos(i_c\,-\,\phi)}
\end{split}
<br>
<br>
\begin{equation}
x_3
\,=\,x\,-\,l\,-\,x_4
\,=\,x\,-\,l\,-\,h\tan(i_c\,-\,\phi)\,+\,x_3\tan{\phi}\tan(i_c\,-\,\phi)
\end{equation}
<br>
\begin{equation}
\Downarrow
\end{equation}
<br>
\begin{equation}
x_3
\,=\,\frac{x\,-\,l\,-h\tan(i_c\,-\,\phi)}{1\,-\,\tan{\phi}\tan(i_c\,-\,\phi)}
\end{equation}
<br>
<br>
Thus:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{h}{V_1\cos{i_c}}
\,+\,\frac{l}{V_2}\,-\,\frac{hV_1}{V_2\sqrt{{V_2}^2\,-\,{V_1}^2}}
\,+\,\frac{x_3}{V_2\cos{\phi}}
\,+\,\frac{h\,-\,x_3\tan{\phi}}{V_1\cos(i_c\,-\,\phi)}
\end{equation}
<br>
<br>
Therefore:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{V_2}{V_1}
\left[\frac{h}{\alpha}\,+\,\frac{h\,-\,x_3\tan{\phi}}{\alpha\cos{\phi}\,-\,V_1\sin{\phi}}\right]
\,+\,
\frac{1}{V_2}\left[l\,-\,\frac{hV_1}{\alpha}\,+\,\frac{x_3}{\cos{\phi}}\right],
\end{equation}
<br>
where
<br>
<br>
\begin{equation}
x_3
\,=\,\frac{x\,-\,l\,-h\beta}{1\,-\,\tan{\phi}\beta}
\text{,}\qquad
\alpha
\,=\,\sqrt{{V_2}^2\,-\,{V_1}^2}
\qquad\text{e}\qquad
\beta
\,=\,\frac{V_1\cos{\phi}\,-\,\alpha\sin{\phi}}{\alpha\cos{\phi}\,-\,V_1\sin{\phi}}
\end{equation}
<br>

```python
import math
import numpy as np
```

```python
def directWave (x, v1):

  return x / v1

def x_1 (h, v1, v2):

  return h * v1 / math.sqrt (v2 ** 2 - v1 ** 2)

def x_1Phi (h, v1, v2, phi):

  return h * math.tan (math.sin (v1 / v2) \
                       - math.radians (phi))

def headWaveA (x, h, v1, v2):

  return x / v2 + 2 * h * math.sqrt (v2 ** 2 - v1 ** 2) / (v1 * v2)

def headWaveB (x, h, v1, v2, l, phi):
  
  phi_rad  = np.radians (phi)
  sin_phi = math.sin (phi_rad)
  cos_phi = math.cos (phi_rad)
  tan_phi = sin_phi / cos_phi

  alpha   = math.sqrt (v2 ** 2 - v1 ** 2)
  beta    = (v1 * cos_phi - alpha * sin_phi) \
          / (alpha * cos_phi - v1 * sin_phi)
  x3      = (x - l - h * beta) / (1 - tan_phi * beta)

  k1 = h / alpha + (h - x3 * tan_phi) \
     / (alpha * cos_phi - v1 * sin_phi)
  k2 = l - h * v1 / alpha + x3 / cos_phi

  return v2 * k1 / v1 + k2 / v2
```

```python
if __name__ == '__main__':

# Modify the parameters and re-execute the code to see how the
# travel-time curves change (tip: try v2 = v3 and v2 < v3)

  xmin = 0   # (km)
  xmax = 20  # (km)
  h    = 2.0 # (km)
  v1   = 1.7 # (km/s)
  v2   = 4.0 # (km/s)
  phi  = 30  # (degrees)
```

```python
if __name__ == '__main__':

  plt.figure (figsize = (16, 8), dpi = 90)

  x1     = x_1 (h, v1, v2)
  x1_phi = x_1Phi (h, v1, v2, phi)
  xc     = 2 * x1

  l = 0.5 * (xmax - xmin)

  xamin = xc
  xamax = l + x1
  xbmin = l + x1_phi
  xbmax = xmax

  x   = np.linspace (xmin, xmax)
  xa  = np.linspace (xamin, xamax)
  xb  = np.linspace (xbmin, xbmax)
  t1  = directWave (x, v1)
  t2a = headWaveA (xa, h, v1, v2)
  t2b = headWaveB (xb, h, v1, v2, l, phi)

  plt.plot (x, t1,
            linewidth = 3,
            label = 'Direct wave')
  plt.plot (xa, t2a,
            color = 'darkorange',
            linewidth = 3,
            label = 'Head wave')
  plt.plot (xb, t2b,
            color = 'darkorange',
            linewidth = 3)

  plt.title ('Travel-time curves',
             fontsize = 28,
             y = 1.05)
  plt.xlabel ('Distance from source (km)',
              fontsize = 20,
              labelpad = 10)
  plt.ylabel ('Travel time (s)',
              fontsize = 20,
              labelpad = 10)
  plt.tick_params (axis = 'x', labelsize = 16)
  plt.tick_params (axis = 'y', labelsize = 16)
  plt.legend (loc = 2, fontsize = 18)
  plt.grid ()
  plt.show ()
```
