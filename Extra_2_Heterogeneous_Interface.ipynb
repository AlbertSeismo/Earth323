###Direct wave

\begin{equation}
t_1\,=\,\frac{x}{V_1}
\end{equation}

###Refraction

For $x\,<\,x_c$ (critical distance), there is no head wave:
<br>
<br>
\begin{equation}
\,x_c
\,=\,\frac{2hV_1}{\sqrt{{V_2}^2\,-\,{V_1}^2}}
\end{equation}
<br>
<br>
When $x_c\,\leq\,x\,\leq\,x_c\,+\,x_2$:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{2\sqrt{{x_1}^2\,+\,h^2}}{V_1}
\,+\,\frac{x\,-\,2x_1}{V_2}
\end{equation}
<br>
<br>
Thus:
<br>
<br>
\begin{equation}
\sqrt{{x_1}^2\,+\,h^2}
\,=\,\frac{h}{\cos{i_{c_1}}}
\,\implies\,t_2
\,=\,\frac{2h}{V_1\cos{i_{c_1}}}
\,+\,\frac{x}{V_2}
\,-\,\frac{2h\sin{i_{c_1}}}{V_2\cos{i_{c_1}}}
\end{equation}
<br>
<br>
Therefore:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{x}{V_2}
\,+\,\frac{2h}{V_1}\left[\frac{1\,-\,\displaystyle\frac{V_1}{V_2}\sin{i_{c_1}}}{\cos{i_{c_1}}}\right]
\end{equation}
<br>
\begin{equation}
1\,-\,\frac{V_1}{V_2}\sin{i_{c_1}}\,=\,{\cos{i_{c_1}}}^2
\,\implies\,t_2
\,=\,\frac{x}{V_2}
\,+\,\frac{2h\cos{i_{c_1}}}{V_1}
\end{equation}
<br>
<br>
Hence:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{x}{V_2}
\,+\,\frac{2h\sqrt{{V_2}^2-{V_1}^2}}{V_1V_2}
\end{equation}
<br>
<br>
Being $l\,=\,x_1\,+\,x_2$, for $x\,\geq\,l\,+\,x_4$:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{\sqrt{{x_1}^2\,+\,h^2}}{V_1}
\,+\,\frac{x_2}{V_2}
\,+\,\frac{x_3}{V_3}
\,+\,\frac{\sqrt{{x_4}^2\,+\,h^2}}{V_1}
\end{equation}
Thus:
<br>
<br>
\begin{equation}
t_2
\,=\,\frac{hV_2}{V_1\sqrt{{V_2}^2\,-\,{V_1}^2}}
\,+\,\frac{l}{V_2}
\,-\,\frac{hV_1}{V_2\sqrt{{V_2}^2\,-\,{V_1}^2}}
\,+\,\frac{x_3}{V_3}
\,+\,\frac{hV_3}{V_1\sqrt{{V_3}^2\,-\,{V_1}^2}}
\end{equation}
<br>
<br>
\begin{equation}
x_3
\,=\,x\,-\,l\,-\,x_4
\qquad\text{e}\qquad
x_4
\,=\,h\tan{i_{c_2}}\,=\,\frac{hV_1}{\sqrt{{V_3}^2\,-\,{V_1}^2}}
\end{equation}
<br>
\begin{equation}
\Downarrow
\end{equation}
<br>
\begin{equation}
t_2
\,=\,\frac{hV_2}{V_1\sqrt{{V_2}^2\,-\,{V_1}^2}}
\,+\,\frac{l}{V_2}
\,-\,\frac{hV_1}{V_2\sqrt{{V_2}^2\,-\,{V_1}^2}}
\,+\,\frac{x}{V_3}
\,-\,\frac{l}{V_3}
\,-\,\frac{hV_1}{V_3\sqrt{{V_3}^2\,-\,{V_1}^2}}
\,+\,\frac{hV_3}{V_1\sqrt{{V_3}^2\,-\,{V_1}^2}}
\end{equation}
<br>
<br>
Therefore:
<br>
<br>
\begin{equation}
t_2
\,=\,\,\frac{x}{V_3}
\,+\,l\left(\frac{1}{V_2}\,-\,\frac{1}{V_3}\right)
\,+\,\frac{h}{\sqrt{{V_2}^2\,-\,{V_1}^2}}\left(\frac{V_2}{V_1}\,-\,\frac{V_1}{V_2}\right)
\,+\,\frac{h}{\sqrt{{V_3}^2\,-\,{V_1}^2}}\left(\frac{V_3}{V_1}\,-\,\frac{V_1}{V_3}\right)
\end{equation}

```python
import math
import numpy as np
```

```python
def directWave (x, v1):

  return x / v1

def x_1 (v1, v2, h):

  return h * v1 / math.sqrt (v2 ** 2 - v1 ** 2)

def x_4 (v1, v3, h):

  return h * v1 / math.sqrt (v3 ** 2 - v1 ** 2)

def headWaveA (x, h, v1, v2):

  return x / v2 + 2 * h * math.sqrt (v2 ** 2 - v1 ** 2) / (v1 * v2)

def headWaveB (x, h, v1, v2, v3, l):

  return x / v3 + l * (1.0 / v2 - 1.0 / v3) \
         + h * (v2 / v1 - v1 / v2) / math.sqrt (v2 ** 2 - v1 ** 2) \
         + h * (v3 / v1 - v1 / v3) / math.sqrt (v3 ** 2 - v1 ** 2)
```

```python
if __name__ == '__main__':

# Modify the parameters and re-execute the code to see how the
# travel-time curves change (tip: try v2 = v3 and v2 < v3)

  xmin = 0   # (km)
  xmax = 20  # (km)
  h    = 2.0 # (km)
  v1   = 1.7 # (km/s)
  v2   = 6.0 # (km/s)
  v3   = 2.5 # (km/s)
```

```python
if __name__ == '__main__':

  plt.figure (figsize = (16, 8), dpi = 100)

  x1 = x_1 (v1, v2, h)
  x4 = x_4 (v1, v3, h)
  xc = 2 * x1

  l = 0.5 * (xmax - xmin)

  xamin = xc
  xamax = l + x1
  xbmin = l + x4
  xbmax = xmax

  x   = np.linspace (xmin, xmax)
  xa  = np.linspace (xamin, xamax)
  xb  = np.linspace (xbmin, xbmax)
  t1  = directWave (x, v1)
  t2a = headWaveA (xa, h, v1, v2)
  t2b = headWaveB (xb, h, v1, v2, v3, l)

  plt.plot (x, t1,
            linewidth = 3,
            label = 'Direct wave')
  plt.plot (xa, t2a,
            color = 'darkorange',
            linewidth = 3,
            label = 'Head wave')
  plt.plot (xb, t2b,
            color = 'darkorange',
            linewidth = 3)

  plt.title ('Travel-time curves',
             fontsize = 28,
             y = 1.05)
  plt.xlabel ('Distance from source (km)',
              fontsize = 20,
              labelpad = 10)
  plt.ylabel ('Travel time (s)',
              fontsize = 20,
              labelpad = 10)
  plt.tick_params (axis = 'x', labelsize = 18)
  plt.tick_params (axis = 'y', labelsize = 18)
  plt.legend (loc = 2, fontsize = 18)
  plt.grid ()
  plt.show ()
```
