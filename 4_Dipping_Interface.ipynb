###Direct wave

\begin{equation}
t_1\,=\,\frac{x}{V_1}
\end{equation}

###Head wave

\begin{equation}
t_2\,=\,\frac{X}{V_{ap}}\,+\,\frac{2h_a\cos{i_c}}{V_1}
\,=\,\frac{X}{V_{ap}}\,+\,\frac{2h_a\sqrt{{V_2}^2-{V_1}^2}}{V_1V_2},\quad\text{where}
\quad V_{ap}\,=\,\frac{V_1}{\sin(i_c + \phi)}
\end{equation}
<br>

####Critical distance

\begin{equation}
x_c
\,=\,2h\frac{\tan{i_c}}{\cos{\phi}}
\,\implies\,x_c
\,=\,\frac{2hV_1}{\cos{\phi}\sqrt{{V_2}^2\,-\,{V_1}^2}}
\end{equation}

```python
import math
import numpy as np
```

```python
def directWave (x, v1):

  return x / v1

def criticalDistance (ha, v1, v2, phi):

  return 2 * ha * v1 / (math.cos (math.radians (phi)) \
                        * math.sqrt (v2 ** 2 - v1 ** 2))

def headWave (x, ha, v1, v2, phi):

  vap = v1 / math.sin (math.asin (v1 / v2) \
                       + math.radians (phi))
  
  return x / vap  + 2 * ha * math.sqrt (v2 ** 2 - v1 ** 2) \
                           / (v1 * v2)
```

```python
if __name__ == '__main__':

# Modify the parameters and re-execute the code to see
# how the travel-time curves change

  xmin = 0   # (km)
  xmax = 20  # (km)
  ha   = 2.0 # (km)
  phi  = 10  # (degrees)
  v1   = 1.7 # (km/s)
  v2   = 4.0 # (km/s)
```

```python
 if __name__ == '__main__':

  plt.figure (figsize = (17, 9), dpi = 70)

  xc = criticalDistance (ha, v1, v2, phi)

  x1 = np.linspace (xmin, xmax)
  x2 = np.linspace (xc, xmax)

  t1 = directWave (x1, v1)
  t2 = headWave (x2, ha, v1, v2, phi)

  plt.plot (x1, t1,
            linewidth = 3,
            label = 'Direct wave')
  plt.plot (x2, t2,
            linewidth = 3,
            label = 'Head wave')

  plt.title ('Travel-time curve',
             fontsize = 30,
             y = 1.05)
  plt.xlabel ('Distance from the source (km)',
              fontsize = 22,
              labelpad = 10)
  plt.ylabel ('Travel time (s)',
              fontsize = 22,
              labelpad = 10)
  plt.tick_params (axis = 'x', labelsize = 20)
  plt.tick_params (axis = 'y', labelsize = 20)
  plt.legend (loc = 2, fontsize = 20)
  plt.grid ()
  plt.show ()
```
