###Direct wave

\begin{equation}
t_1\,=\,\frac{x}{V_1}
\end{equation}

###Head wave

\begin{equation}
t_2
\,=\,\frac{x}{V_2}\,+\,\frac{2h\cos{i_c}}{V_1}
\,=\,\frac{x}{V_2}\,+\,\frac{2h\sqrt{{V_2}^2-{V_1}^2}}{V_1V_2}
\end{equation}



####Critical distance

\begin{equation}
x_c
\,=\,2h\tan{i_c}
\end{equation}
<br>
\begin{equation}
\sin{i_c}
\,=\,\frac{V_1}{V_2}
\quad\text{and}\quad\cos{i_c}
\,=\,\sqrt{1\,-\,{\sin{i_c}}^2}
\,\implies\,\cos{i_c}
\,=\,\sqrt{1\,-\,\frac{{V_1}^2}{{V_2}^2}}
\end{equation}
<br>
\begin{equation}
\tan{i_c}
\,=\,\frac{\sin{i_c}}{\cos{i_c}}
\,=\,\frac{\displaystyle\frac{V_1}{V_2}}{\displaystyle\sqrt{1\,-\,\frac{{V_1}^2}{{V_2}^2}}}
\,=\,\frac{1}{\displaystyle\sqrt{\frac{{V_2}^2}{{V_1}^2}\,-\,1}}
\,=\,\frac{1}{\displaystyle\sqrt{\frac{{V_2}^2\,-\,{V_1}^2}{{V_1}^2}}}
\,=\,\frac{V_1}{\sqrt{{V_2}^2\,-\,{V_1}^2}}
\end{equation}
<br>
Therefore:
\begin{equation}
x_c
\,=\,\frac{2hV_1}{\sqrt{{V_2}^2\,-\,{V_1}^2}}
\end{equation}

###Reflection

\begin{equation}
t_{r_1}\,=\,\frac{2\sqrt{\displaystyle\frac{x^2}{4}\,+\,h^2}}{V_1}
\end{equation}

```python
import math
import numpy as np
```

```python
def directWave (x, v1):

  return x / v1

def criticalDistance (h, v1, v2):

  return 2 * h * v1 / math.sqrt (v2 ** 2 - v1 ** 2)

def headWave (x, h, v1, v2):

  return x / v2 + 2 * h * math.sqrt (v2 ** 2 - v1 ** 2) / (v1 * v2)

def reflectedWave (x, h, v1):
  
  return 2 * np.sqrt (0.25 * x ** 2 + h ** 2) / v1
```

```python
if __name__ == '__main__':

# Modify the parameters and re-execute the code to see
# how the travel-time curves change

  xmin = 0   # (km)
  xmax = 20  # (km)
  h    = 2.0 # (km)
  v1   = 1.7 # (km/s)
  v2   = 4.0 # (km/s)
```

```python
 if __name__ == '__main__':

  plt.figure (figsize = (16, 8), dpi = 90)

  xc = criticalDistance (h, v1, v2)

  x1 = np.linspace (xmin, xmax)
  x2 = np.linspace (xc, xmax)

  t1  = directWave (x1, v1)
  t2  = headWave (x2, h, v1, v2)
  tr1 = reflectedWave (x1, h, v1)

  plt.plot (x1, t1,
            linewidth = 3,
            label = 'Direct wave')
  plt.plot (x2, t2,
            linewidth = 3,
            label = 'Head wave')
  plt.plot (x1, tr1,
            linewidth = 3,
            label = 'Reflected wave')

  plt.title ('Travel-time curves',
             fontsize = 28,
             y = 1.05)
  plt.xlabel ('Distance from the source (km)',
              fontsize = 22,
              labelpad = 10)
  plt.ylabel ('Travel time (s)',
              fontsize = 22,
              labelpad = 10)
  plt.tick_params (axis = 'x', labelsize = 16)
  plt.tick_params (axis = 'y', labelsize = 16)
  plt.legend (loc = 2, fontsize = 18)
  plt.grid ()
  plt.show ()
```
